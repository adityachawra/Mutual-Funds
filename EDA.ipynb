```python

new_df = []

for date in df.keys():
    for ele in df[date]:
        new_df.append({'date':date,'scheme_code': ele['scheme_code'],'nav': ele['nav']})
       


        
```

```python
nav_df = pd.DataFrame(new_df,columns=['date', 'scheme_code', 'nav']) # Creating a new DataFrame with the desired columns
nav_df.head()



```

```python
pivot_df = nav_df.pivot(index='date', columns='scheme_code', values='nav') # Pivoting the DataFrame
pivot_df = pivot_df.reset_index() # Resetting the index

a = pd.merge(date_df, pivot_df, on='date', how='left')
a.head()


            
```

```python
fund_specs = {}
for date in df.keys():
    for ele in df[date]:
        if ele['scheme_code'] not in fund_specs:
            fund_specs[ele['scheme_code']] = {'shceme_name':ele['scheme_name'], 'category':ele['category'], 'amc':ele['amc']}
```

```python
fund_specs_df = pd.DataFrame(fund_specs)
fund_specs_df = fund_specs_df.transpose() # Transposing the DataFrame
fund_specs_df = fund_specs_df.reset_index() # Resetting the index
fund_specs_df.columns = ['scheme_code', 'scheme_name', 'category', 'amc'] # Renaming the columns
fund_specs_df.head()
```

```python
Mutual_funds_specs = fund_specs_df
```

```python
import json
with open('Mutual_funds_specs.json', 'w') as f:
    json.dump(fund_specs, f, indent=2)  # indent=2 makes it human-readable
```
